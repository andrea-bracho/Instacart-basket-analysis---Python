# Question 5:
## Determine a suitable way to combine the orders_products_combined dataframe with your products data set. 

```python
# Look at orders_products_combined
df_ords_prods.head()
```

```python
# Remove unncessary columns and assign to new df
df_ords_prods_1 = df_ords_prods.drop(columns=['Unnamed: 0', '_merge'])
df_ords_prods_1.head()
```

```python
# Look at "products_checked'
df_prods.head()
```

```python
# Remove unncessary columns and assign to new df
df_prods_1 = df_prods.drop(columns=['Unnamed: 0'])
df_prods_1.head()
```

```python
# They share the Product_id column, so I will use that one to create a LEFT merge. 
df_prods_ords_merge = df_ords_prods_1.merge(df_prods_1, on = 'product_id', how = 'left', indicator = True)
```
