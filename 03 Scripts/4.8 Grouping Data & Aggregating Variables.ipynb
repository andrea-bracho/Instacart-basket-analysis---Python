## Question 4:
### Create a loyalty flag for existing customers. Use the transform() & loc() functions.

```python
# New column "max_order". Shows the max mumber of order per customer. 
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(20)
```

```python
# Create a column 'loyalty_flag' indicating loyalty status of customer according to the number of orders placed.
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Look at the count of each category in 'loyalty_flag'
df_ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
# So we can see more rows using .head()
pd.options.display.max_rows = None
# Prints only columns that i am interested on
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number', 'max_order']].head(100)
```

## Question 6:
### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
#### * If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#### * If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# New column "avg_price". Shows the average price paid by each customer. 
df_ords_prods_merge['avg_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create a column 'spending_flag' indicating loyalty status of customer according to the number of orders placed.
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
# Get counts of spending_flag types
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge[['order_id', 'user_id', 'max_order', 'loyalty_flag', 'spending_flag']].head(100)
```

## Question 7:
### In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
#### * If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#### * If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#### * If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# New column "Frequency_of_orders". Shows the median value of each customer's 'days_since_prior_order' 
df_ords_prods_merge['frequency_of_orders'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.head(200)
```

```python
# Create a column 'order_frequency_flag' indicating loyalty status of customer according to the number of orders placed.
df_ords_prods_merge.loc[df_ords_prods_merge['frequency_of_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['frequency_of_orders'] > 10) & (df_ords_prods_merge['frequency_of_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['frequency_of_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Check table 
df_ords_prods_merge[['order_id', 'user_id', 'max_order', 'loyalty_flag', 'spending_flag', 'frequency_of_orders', 'order_frequency_flag']].tail(200)
```

```python
# Get counts of order_frequency_flag types
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```
