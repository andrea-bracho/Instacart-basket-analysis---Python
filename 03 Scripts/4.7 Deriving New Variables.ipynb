## 2. if-else statements

```python
# Create new dataframe to use only the first million rows of data
df = df_ords_prods_merged[:1000000]
```

```python
# Define the function 
def price_label(row):

# How it will run
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df_ords_prods_merged.head()
```

```python
# Apply the function 'price_label' into our df.   Axis=1 --> ROWS.    Axis=0 --> COLUMNS
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# check our price labels
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
df.head()
```

## 3. IF  using the loc() function

### name.loc[name['column'] <>= # , 'new_name_loc'] = outcome
### if , then
### if dealing with multiple conditions, write them in (cond1) & (cond2)

```python
df.loc[df['prices'] > 15, 'price_label'] = 'High-range product'
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_label'] = 'Mid-range product' 
df.loc[df['prices'] <= 5, 'price_label'] = 'Low-range product'
```

```python
# Getting counts. Exactly the same as previous method
df['price_label'].value_counts(dropna = False)
```

```python
# Using conditions on entire dataframe:
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_label'] = 'High-range product'
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df['prices'] > 5), 'price_label'] = 'Mid-range product' 
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_label'] = 'Low-range product'
```

```python
# Counting all types of product:
df_ords_prods_merged['price_label'].value_counts(dropna = False)
```

## 4. if statements with for-loops

```python
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

#### Saturday (0) is the busiest day
#### Wednesday (4) is the least busy day

```python
#Create a for-loop:

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Print 
result
```

```python
#add column
df_ords_prods_merged['busiest_day']=result
```

```python
# Get counts of days according to number of orders
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

## Question 2:
### Suppose your clients have changed their minds about the labels you created in your busiest_day column. Now, they want Busiest day to become Busiest days (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, theyd also like to know the two slowest days. Create a new column for this using a suitable method

```python
# Get the counts of each day Sat(0), Sun(1), Mon(2), Tues(3), Wed(4), Thur(5), Fri(6)
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create an if statetment using the .loc() function
# 0 and 1 represent the busiest days (Saturday and Sunday)
# 3 and 4 represent the slowest days (Tuesday & Wednesday)
# 2,5,6 represent regular days (Monday, Thursday, and Friday)

df_ords_prods_merged.loc[(df_ords_prods_merged['orders_day_of_week'] == 0) | (df_ords_prods_merged['orders_day_of_week'] == 1), 'Busy_days_by_orders'] = 'Busiest days'
df_ords_prods_merged.loc[(df_ords_prods_merged['orders_day_of_week'] == 3) | (df_ords_prods_merged['orders_day_of_week'] == 4), 'Busy_days_by_orders'] = 'Slowest days'
df_ords_prods_merged.loc[(df_ords_prods_merged['orders_day_of_week'] == 2) | (df_ords_prods_merged['orders_day_of_week'] == 5) | (df_ords_prods_merged['orders_day_of_week'] == 6), 'Busy_days_by_orders'] = 'Regularly days'
```
